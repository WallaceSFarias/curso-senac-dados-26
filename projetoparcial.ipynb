```python
import pandas as pd
import numpy as np
```

```python
dados_semestre_1="Preços semestrais - AUTOMOTIVOS_2025.01.csv"

df_semestre_1=pd.read_csv(dados_semestre_1, sep = ";" , encoding ='latin1')

df_semestre_1
```

```python
dados_semestre_2="Preços semestrais - AUTOMOTIVOS_2025.02.csv"

df_semestre_2=pd.read_csv(dados_semestre_2, sep = ";" , encoding ='latin1')

df_semestre_2
```

```python
df_semestre_1.info()
```

```python
df_semestre_2.info()
```

```python
df_semestre_2['CNPJ da Revenda'] = df_semestre_2['CNPJ da Revenda'].astype(str)
```

```python
df_anp=pd.concat([df_semestre_1,df_semestre_2], ignore_index = True)
df_anp
```

```python
df_anp['Valor de Venda']= df_anp['Valor de Venda'].str.replace(',','.').astype(float)
df_anp
```

```python
df_anp.info()
```

```python
df_gasolina=df_anp.loc[df_anp['Produto']=='GASOLINA']
df_etanol=df_anp.loc[df_anp['Produto']=='ETANOL']
```

```python
df_gasolina
```

```python
df_etanol
```

```python
df_media_est_gaso = df_gasolina.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index().sort_values(by='Valor de Venda',ascending=False)

df_media_est_gaso.head()
```

```python
df_media_est_etanol = df_etanol.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index().sort_values(by='Valor de Venda',ascending=False)

df_media_est_etanol.head()
```

```python
# Comparação Regional de Preços
df_media_reg_gaso=df_gasolina.groupby('ï»¿Regiao - Sigla')['Valor de Venda'].mean().reset_index().sort_values(by='Valor de Venda',ascending=False)

df_media_reg_gaso
```

```python
df_media_reg_etanol = df_etanol.groupby('ï»¿Regiao - Sigla')['Valor de Venda'].mean().reset_index().sort_values(by='Valor de Venda',ascending=False)

df_media_reg_etanol
```

```python
df_dez_maiores_munic_gasol=df_gasolina.groupby('Municipio')['Valor de Venda'].mean().reset_index().sort_values(by='Valor de Venda',ascending=False)

df_dez_maiores_munic_gasol.head(10)
```

```python
df_dez_maiores_munic_etanol=df_etanol.groupby('Municipio')['Valor de Venda'].mean().reset_index().sort_values(by='Valor de Venda',ascending=False)

df_dez_maiores_munic_etanol.head(10)
```
